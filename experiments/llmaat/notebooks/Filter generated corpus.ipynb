```python
import pandas as pd
```

```python
ROOT_DIR = ""
```

```python
df_en = pd.read_csv(f'{ROOT_DIR}/qe-reranked/diverse_sample.1M.en', header=None, sep='\t')
```

```python
with open(f'{ROOT_DIR}/en-ru/diverse_sample.1M.en') as f:
    en_lines = f.read().split('\n')
with open(f'{ROOT_DIR}/en-ru/diverse_sample.1M.ru_RU') as f:
    ru_lines = f.read().split('\n')
with open(f'{ROOT_DIR}/en-ru/diverse_sample.1M.scores') as f:
    scores_lines = [float(x) for x in f.read().split('\n')]
```

```python
end_of_turns = ['<end_of_turn>' in line for line in ru_lines]
```

```python
df = pd.DataFrame({'en': en_lines, 'ru': ru_lines, 'score': scores_lines, 'has_end_of_turn': end_of_turns})
```

```python
len(df[df.has_end_of_turn == False])
```

```python
df.score.hist(bins=50)
```

```python
def printdf(df):
    for i,v in df.iterrows():
        print(v.score)
        print(v.en)
        print(v.ru)
        print()
    
printdf(df[df.has_end_of_turn == False].sort_values(by='score').tail(10))
```

```python
    
printdf(df[(df.has_end_of_turn == False) & (df.score < 2)].sort_values(by='score').tail(20))
```

```python
df_filtered = df[(df.has_end_of_turn == False) & (df.score < 2)]
len(df_filtered)
```

```python
df_filtered
```

```python
df_filtered.en.to_csv(f'{ROOT_DIR}/qe-reranked/diverse_sample.1M.filtered2.en', index=False, header=None)
df_filtered.ru.to_csv(f'{ROOT_DIR}/qe-reranked/diverse_sample.1M.filtered2.ru', index=False, header=None)
```

# ZH

```python
with open(f'{ROOT_DIR}/diverse_sample.1M.en') as f:
    en_lines = f.read().split('\n')
with open(f'{ROOT_DIR}/diverse_sample.1M.zh_CN') as f:
    zh_lines = f.read().split('\n')
with open(f'{ROOT_DIR}/diverse_sample.1M.scores') as f:
    scores_lines = [float(x) for x in f.read().split('\n')]

df = pd.DataFrame({'en': en_lines, 'zh': zh_lines, 'score': scores_lines})
```

```python
printdf(df[df.score <2].sort_values(by='score').tail(10))
```
